```python
# Complete the functions defined in this cell


# Test 01: Play Right
def playRight(sequence,beats):
  n=0
  for i in beats:
    if i==1:
      n+=1
  for j in range(n):
    k=sequence[len(sequence)-1]
    for i in range(len(sequence)-2,-1,-1):
      sequence[i+1]=sequence[i]
    sequence[0]=k

  return sequence


# Test 02: Discard Cards
def discardCards(cards,number):
  count=0
  for i in cards:
    if i==number:
      count+=1

  for j in range(count):
    index=0
    for i in cards:
      if i!=number:
        index+=1
      elif i==number:
        break

    for i in range(index,len(cards)-1):
      cards[i]=cards[i+1]
    cards[len(cards)-1]=0

  return cards


#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  j=0
  merge=[]
  for i,char in enumerate(pokemon_1):
    if char==None:
      pokemon_1[i]=0
  for i,char in enumerate(pokemon_2):
    if char==None:
      pokemon_2[i]=0
  while j<len(pokemon_1):
    merge+=[pokemon_1[j]+pokemon_2[len(pokemon_2)-1-j]]
    j+=1
  return np.array(merge)


# Test 04: Balance your Salami
def balanceSalami(salami):
  for n in range(2,len(salami)-1):
    sum1=0
    sum2=0
    l1=[]
    l2=[]
    for i in range(0,n):

      l1+=[salami[i]]
    for j in range(n,len(salami)):
      l2+=[salami[j]]
    for i in l1:
      sum1+=i
    for i in l2:
      sum2+=i
    if sum1==sum2:
      ans=True
      break
    else:
      ans=False
  return ans

# Test 05: Protecc Salami
def protectSalami(salami):
  d={}
  l=[]

  for i in salami:
    if i not in l:
      l+=[i]
  for j in l:
    k=0
    for n in salami:
      if j==n:
        k+=1
    d[j]=k
  for m in d:
    c=0
    for p in d:

        if d[p]==d[m] and d[p]!=1:
            c+=1
        if c>=2:
            ans=True
            return ans
            break

        else:
            ans=False
  return ans


# Test 05: Odd Even Wave
def waveYourFlag(arr):
  for i in range(-1,len(arr)-1):
    if (arr[i-1]%2==0 and arr[i]%2==0 ) or (arr[i-1]%2!=0 and arr[i]%2!=0 ):
      c=arr[i]
      arr[i]=arr[i+1]
      arr[i+1]=c


  return arr




```
